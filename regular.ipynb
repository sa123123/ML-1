```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
np.random.seed(42)
n_sample=100
x=np.random.randn(n_sample)*10
y=x*0.5+np.random.randn(n_sample)*2
model.fit(x.reshape(-1,1),y)
pred=model.predict(x.reshape(-1,1))
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x,y)
plt.plot(x,pred)
plt.show()
```

```python
plt.figure(figsize=(10,8));
plt.scatter(x, y);

o_lr = LinearRegression(normalize=True)
o_lr.fit(x.reshape(-1, 1), y)
o_model_pred = o_lr.predict(x.reshape(-1,1))

plt.scatter(x, y);
plt.plot(x, o_model_pred);
print("Coefficient Estimate: ", o_lr.coef_)
```

```python
idx=x.argmax()
y[idx]=200
idx=x.argmin()
y[idx]=-200
```

```python
plt.figure(figsize=(10,8));
plt.scatter(x, y);

o_lr = LinearRegression(normalize=True)
o_lr.fit(x.reshape(-1, 1), y)
o_model_pred = o_lr.predict(x.reshape(-1,1))

plt.scatter(x, y);
plt.plot(x, o_model_pred);
print("Coefficient Estimate: ", o_lr.coef_)
```

# Ridge

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=0.5,normalize=True)
ridge.fit(x.reshape(-1,1),y)
ridge_model_pred=ridge.predict(x.reshape(-1,1))
plt.figure(figsize=(10,8));
plt.scatter(x, y);
plt.plot(x, ridge_model_pred);
print("Coefficient Estimate: ", ridge.coef_)
```

# LASSO

```python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=0.4,normalize=True)
lasso.fit(x.reshape(-1,1),y)
lasso_model_pred=lasso.predict(x.reshape(-1,1))
plt.figure(figsize=(10,8));
plt.scatter(x, y);
plt.plot(x, lasso_model_pred);
print("Coefficient Estimate: ", lasso.coef_)
```

# elastic net

```python
from sklearn.linear_model import ElasticNet
en=ElasticNet(alpha=0.4,normalize=True)
en.fit(x.reshape(-1,1),y)
en_model_pred=en.predict(x.reshape(-1,1))
plt.figure(figsize=(10,8));
plt.scatter(x, y);
plt.plot(x, en_model_pred);
print("Coefficient Estimate: ", en.coef_)
```
